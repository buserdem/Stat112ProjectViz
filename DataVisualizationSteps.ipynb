```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
sns.color_palette("colorblind")
sns.set_style("darkgrid")
sns.set_context("notebook")
#All necessary libraries were imported, and color palette and grid were assigned
#Color palette was chosen consciously for color blinds
```

```python
book_price=pd.read_excel('book_priceCleanData.xlsx')
book_price.head()
#The data were observed
```

```python
book_price.corr()
#Correlation coefficient between binary variables
```

```python
plt.figure(figsize = (8,8))  
sns.boxplot(data=book_price, x="Rating", y="Type")
plt.title('Relationship Between Rating and Type',fontsize = 15, fontweight = "bold")
plt.xlabel('Rating')
plt.ylabel('Book Types')
plt.show() 
#One categoric and one numeric variable visualized with customized box plot
```

```python

plt.figure(figsize = (8,8))  
sns.relplot(data=book_price, y="Price", x="Clicking Rate", kind='scatter' )
plt.title('The Association Between Price and Clicking Rate',fontsize = 15, fontweight = "bold")
plt.xlabel('Clicking Rate')
plt.ylabel('Price')
plt.show() 
#Two numeric varible were visualized with customized scatterplot 
```

```python
plt.figure(figsize = (8,8)) 
sns.catplot(data=book_price, y="Price", x="Second Hand",
             kind='bar') 
plt.title('Association Between Price and Second Hand',fontsize = 15, fontweight = "bold")
plt.xlabel('Second Hand')
plt.ylabel('Price')
plt.show()   
#One categoric and one numeric variable were visualized with customized bar plot
```

```python
plt.figure(figsize = (8,8)) 
sns.violinplot(data=book_price, x="Type", y="Price") 
plt.title('The Relationship between Type and Price',fontsize = 15, fontweight = "bold")
plt.xlabel('Type')
plt.ylabel('Price')
plt.show()  
#One categoric and one numeric varible visualized with customized violin plot in order to observe the density also
```

```python
plt.figure(figsize = (8,8))  
sns.violinplot(data=book_price, x="Publisher", y="Price")
plt.title('The Relationship Between Publisher and Price',fontsize = 15, fontweight = "bold")
plt.xlabel('Publisher')
plt.ylabel('Price')
plt.show()  
#One categoric and one numeric varible visualized with customized violin plot in order to observe the density also
```

```python
plt.figure(figsize = (8,8)) 
sns.boxplot(data=book_price, x="Clicking Rate", y="Type") 
plt.title('The Relationship between Type and Clicking Rate',fontsize = 15,fontweight = "bold")
plt.ylabel('Type')
plt.show() 
#One categoric and one numeric varible were visualized with customized box plot
```

```python
sns.lineplot(x='Year' , y='Clicking Rate' , data=book_price,ci=None) 
plt.title('Clicking Rate Over the Years',fontsize = 15, fontweight = "bold")
#Year varible was visualized with customized time series graph
```

```python
plt.figure(figsize = (8,8)) 
sns.relplot(data=book_price, x="No Of Page", y="Shipping Fee", kind='scatter')
plt.title('Relationship Between Number Of Pages and Shipping Fee',fontsize = 15, fontweight = "bold")
plt.xlabel('No Of Page')
plt.ylabel('Shipping Fee')
plt.show()
#Two numeric variables were visualized with customized scatter plot
```
